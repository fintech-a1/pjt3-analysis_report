```python
# 다음 금융에서 종목별 상장일 Crawling
import time
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException, NoSuchWindowException
from selenium import webdriver

options = webdriver.ChromeOptions()
options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko")
driver = webdriver.Chrome(options=options)
listing_date_lists = []
for e_code in close_error_df['code']:
    uri = 'https://wisefn.finance.daum.net/v1/company/c1020001.aspx?cmp_cd={}'.format(e_code)
    driver.get(uri)
    listing_date=driver.find_element_by_xpath('//*[@id="cTB201"]/tbody/tr[3]/td[1]').text
    listing_date_lists.append(listing_date)
```
