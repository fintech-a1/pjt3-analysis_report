```python
import time
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException, NoSuchWindowException
from selenium import webdriver

uri = 'http://data.krx.co.kr/contents/MDC/MDI/mdiLoader/index.cmd?menuId=MDC0201020103'
options = webdriver.ChromeOptions()
options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko")
driver = webdriver.Chrome(options=options)
driver.get(uri)
```

```python
# 날짜만 바꿔서 KRX에서 크롤링해준다.
for i in nan_df_29.index[13:16]:
    try:
        input_stock = driver.find_element_by_name('tboxisuCd_finder_stkisu0_0')
        input_st_dt = driver.find_element_by_name('strtDd')
        input_ed_dt = driver.find_element_by_name('endDd')
        input_stock.clear()
        input_st_dt.clear()
        input_ed_dt.clear()
        input_stock.send_keys(nan_df_29['code'].loc[i])
        day=nan_df_29['작성일'].loc[i]
        input_st_dt.send_keys(''.join(str(XKRX.previous_open(pd.Timestamp(day)))[:10].split('-')))
        input_ed_dt.send_keys(''.join(str(XKRX.previous_open(pd.Timestamp(day)))[:10].split('-')))
        search_btn_1 = driver.find_element_by_id('btnisuCd_finder_stkisu0_0')
        search_btn_1.click()
        time.sleep(2.4)
        search_btn_2 = driver.find_element_by_id('jsSearchButton')
        search_btn_2.click()
        time.sleep(2.4)
        open_price = driver.find_element_by_xpath('//*[@id="jsMdiContent"]/div/div[1]/div[1]/div[1]/div[2]/div/div/table/tbody/tr/td[5]').text
        close_price = driver.find_element_by_xpath('//*[@id="jsMdiContent"]/div/div[1]/div[1]/div[1]/div[2]/div/div/table/tbody/tr/td[2]').text
        nan_df_29['시가'].loc[i] = open_price
        nan_df_29['종가'].loc[i] = close_price
        no_search = driver.find_element_by_xpath('//*[@id="jsGrid_div__finder_stkisu0_0"]/div/div/div[2]/div').text
        driver.find_element_by_xpath('//*[@id="jsLayer_finder_stkisu0_0"]/div[1]/button/span[1]').click()
        time.sleep(1)

        if no_search :
            nan_df_29['시가'].loc[i] = np.nan
            nan_df_29['종가'].loc[i] = np.nan
            time.sleep(1)
            
    except (NoSuchElementException, StaleElementReferenceException) as e :
        pass
```
